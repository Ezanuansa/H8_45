```python
alice_wordcloud = WordCloud(background_color = 'white',
                            max_words = 2000,
                            stopwords = STOPWORDS)
alice_wordcloud.generate(alice)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(alice_wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
# updated stopwords by addings by word "said"
for word in ["said", "go", "will", "oh"]
STOPWORDS.add(word)

# regenerate
alice_wordcloud.generate(alice)

#vizualized by wordcloud
plt.figure(figsize=(10, 5))
plt.imshow(alice_wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
canada = pd.read_excel("data/Canada.xlsx", sheet_name = "Canada by Citizenship",
                         skiprows=range(20), skipfooter=2)
canada
```

```python
canada.shape
```

```python
canada = canada.drop(columns=["Type", "Coverage", "AREA", "REG", "DEV"])
canada.head()
```

```python
canada = canada.rename(columns={
    "OdName": "Country",
    "RegName": "Region",
    "AreaName": "Continent",
})
canada.head()

canada.set_index("Country", inplace=True)

canada["Total"] = canada.sum(axis=1)
years=list(range(1980, 2014))

canada.head()
```

```python
max_words = 200

word_string = ""
for country in canada.index:
    num_repeat = int(canada.loc[country, "Total"] / float(canada.Total.sum()) * max_words)
    word_string = word_string + ((country + " ") * num_repeat)
    
word_string

```
