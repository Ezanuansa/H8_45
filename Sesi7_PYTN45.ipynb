> Top 5 Countries that cotributed the most immigrants to Canada

```python
top_5_countries = df_canada.sort_values(by="Total", ascending=False).head(5)[years]
top_5_countries
```

```python
top_5_countries.T.plot(figsize = (12,6))
plt.title("Top 5 Countries that cotributed the most immigrants to Canada")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.show()
```

```python
top_5_countries.T.plot(kind='area', 
                       figsize = (14,7),
                      stacked=False,
                      alpha=.35)
plt.title("Top 5 Countries that cotributed the most immigrants to Canada")
plt.xlabel("Years")
plt.ylabel("Number of Immigrants")
plt.savefig("data/top_5_country.png")
plt.show()
```

```python
top_10_countries = df_canada.sort_values(by="Total", ascending=False).head(10)["Total"]
top_10_countries
```

```python
continents = df_canada.groupby("Continent").sum()

pie_colors= ["gold", "yellowgreen", "lightcoral", "lightskyblue", "lightgreen", "pink"]
explode_list= [.1, 0, 0, 0, .1, .1]
continents["Total"].plot(kind='pie', figsize=(20, 10), shadow=True,
                        autopct= '%1.lf%%', explode=explode_list, 
                        colors= pie_colors, pctdistance= 1.15, labels= None)
plt.legend(labels=continents.index, loc="upper left")
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.set_title("Immigrant from Indonesia & Malaysia")
ax1.set_ylabel("Number of Immigrant")
ax1.set_xlabel("Year")
ax2.set_title("Immigrant from China & India")
plt.suptitle("Immigrant Contribution", y=1)

imci.loc[["Indonesia", "Malaysia"]].T.plot(kind= "area", ax=ax1)
imci.loc[["China", "India"]].T.plot(kind= "area", ax=ax2)
```

```python
id_ma = imci.loc[["Indonesia", "Malaysia"]]
cn_in = imci.loc[["China", "India"]]
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,6))
for df, ax in zip ((id_ma, cn_in), axes):
    ax.set_title(f"{' and '.join(df.index.tolist())} Immigrants")
    df.T.plot(kind='line', ax=ax)
```
